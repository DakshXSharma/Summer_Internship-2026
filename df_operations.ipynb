```python
import pandas as pd

one = pd.DataFrame({
    "Name": ["A1", "B1", "C1", "D1"],
    "Subject": ["sub1", "sub2", "sub4", "sub6"],
    "Marks": [78, 90, 80, 50]
}, index = [1, 2, 3, 4])
two = pd.DataFrame({
    "Name": ["B1", "C1", "P1", "O1"],
    "Subject": ["sub1", "sub2", "sub3", "sub4"],
    "Marks": [78, 90, 80, 50]
}, index = [1, 2, 3, 4])
print(one, '\n')
print(two, '\n')

res1 = pd.concat([one, two], keys= ['x', 'y'])
print(res1, '\n')
res2 = pd.concat([one, two], ignore_index= True)
print(res2, '\n')
res3 = pd.concat([one, two], axis= 1)
print(res3)
```

```python
import pandas as pd

left = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["A1", "A2", "A3", "A4"],
    "Subject": ["sub1", "sub2", "sub4", "sub6"]
})
right = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["B1", "B2", "B3", "B4"],
    "Subject": ["sub1", "sub2", "sub3", "sub4"]
})

res = left.merge(right, on= 'Subject', how= 'left')
print(res)
```

```python
import pandas as pd

left = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["A1", "A2", "A3", "A4"],
    "Subject": ["sub1", "sub2", "sub4", "sub6"]
})
right = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["B1", "B2", "B3", "B4"],
    "Subject": ["sub1", "sub2", "sub3", "sub4"]
})

res1 = left.merge(right, on= 'Subject', how= 'outer')
print(res1, '\n')
res2 = left.merge(right, on= 'Subject', how= 'inner')
print(res2, '\n')
res3 = left.merge(right, on= 'Subject', how= 'right')
print(res3)
```

```python
import pandas as pd

left = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["A1", "A2", "A3", "A4"],
    "Subject": ["sub1", "sub2", "sub4", "sub6"]
})
right = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["B1", "B2", "B3", "B4"],
    "Subject": ["sub1", "sub2", "sub3", "sub4"]
})

res = left.join(right, lsuffix= '_Left', rsuffix= '_Right')
print(res)
```
