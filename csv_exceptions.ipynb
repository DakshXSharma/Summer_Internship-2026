```python
import csv

d = [
    ["Name", "Email", "Mobile"],
    ["Abc", "abc@fe.fd", "939435"],
    ["Agsfr", "abfec@fe.fd", "23784"],
    ["Flfw", "vfeffc@fe.fd", "879"],
    ["Qrdtf", "ghsd@fe.fd", "38762847"]
]

with open("data.csv", "w", newline="") as file:
    writer = csv.writer(file)
    
    for row in d:
        writer.writerow(row)
```

```python
d2 = [
    ["ufh", "ndjk@fsfd.sdf", "84973"]
]

with open("data.csv", "a", newline="") as file:
    writer = csv.writer(file)
    
    for row in d2:
        writer.writerow(row)
```

```python
with open("data.csv", "r") as file:
    reader = csv.reader(file)
    
    for row in reader:
        print(row)
```
