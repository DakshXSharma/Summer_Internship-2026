```python
import numpy as np

arr = np.array([1, 2, np.nan, 4, np.nan])

res = np.isnan(arr)
print(res)

arr2 = arr[~np.isnan(arr)]
print(arr2)

arr3 = np.nan_to_num(arr, nan= 1)
print(arr3)
```

```python
# -3x + 2y = 5, x - 2y = 5

arr1 = np.array([[-3, 2], [1, -2]])
arr2 = np.array([5, 5])

res = np.linalg.solve(arr1, arr2)
print(res)
```

```python
# 4x + 2y = 8, 5x + 3y = 9

arr1 = np.array([[4, 2], [5, 3]])
arr2 = np.array([8, 9])

res = np.linalg.solve(arr1, arr2)
print(res)
```
