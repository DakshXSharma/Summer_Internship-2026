```python
def function1():
    print("Eg of a function!")


function1()
```

```python
def function2(*city):
    print("Last city: ", city[-1])
function2("jaipur", "kota", "alwar", "ajmer")
```

```python
def function3(**city):
    print(city)
function3(city2 = "jaipur", city1 = "kota", city4 = "alwar", city3 = "ajmer")
```

```python
def function4(city = "Jaipur"):
    print(city)
function4("New Delhi")
function4()
```

```python
def function5(fruit):
    for x in fruit:
        print(x)
fruits = ['apple', 'banana', 'cherry']
function5(fruits)
```

```python
def dob(n1):
    return n1 * 2
print(dob(10))
```

```python
def f1(n1, /):
    print(n1 * 2)
f1(5) # f1(n1 = 5) -> error!
```

```python
def f2(*, n1):
    print(n1 * 2)
f2(n1 = 5) # f2(5) -> error!
```

```python
def f3(a, b, /, *, c, d):
    print(a + b + c + d)
f3(5, 6, c = 7, d = 8)
```

```python
def r1(k):
    if(k > 0):
        print(k)
        k -= 1
        r1(k)
    else:
        print(k)
r1(5)
```
