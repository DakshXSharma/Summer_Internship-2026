```python
import numpy as np

a = np.array([1, 2, 3, 4, 5])
print("1D Array:\n", a, type(a), a.dtype)

a = np.array([[1, 2, 3, 4, 5], [11, 12, 13, 14, 15]])
print("\n2D Array:\n", a, type(a), a.dtype)
```

```python
import numpy as np

a = np.array([
    1, 2, 3, 4, 5
], dtype= 'S')
print(a, type(a), a.dtype)
```

```python
import numpy as np

a = np.zeros(7)
print(a, '\n')
a = np.ones((7, 2, 3))
print(a)
```

```python
import numpy as np

a = np.ones((7, 2), order= 'F')
print(a)
```

```python
import numpy as np

arr1 = np.linspace(0, 5, num= 10)
print(arr1)
arr2 = np.linspace(0, 5, num= 10, endpoint= False)
print(arr2)
arr3 = np.linspace(0, 5, num= 10, retstep= True)
print(arr3)
```
