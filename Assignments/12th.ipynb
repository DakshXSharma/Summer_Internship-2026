```python
import numpy as np
```

```python
arr1d = np.array([1, 2, 3, 4, 5, 6])
print(arr1d, '\n')

arr2d = arr1d.reshape(2, 3)
print(arr2d)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr, '\n')

print("Shape:", arr.shape)
print("Dimension:", arr.ndim)
print("Data Type:", arr.dtype)
print("Item Size:", arr.itemsize)
```

```python
arr = np.full((3, 3), 9)
print(arr)
```

```python
arr = np.linspace(25, 125, 10)
print(arr)
```

```python
python_list = [10, 20, 30, 40, 50]
print(python_list, type(python_list))

numpy_array = np.array(python_list)
print(numpy_array, type(numpy_array))
```

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr)

reversed_arr = arr[::-1]
print(reversed_arr)
```

```python
arr = np.arange(1, 49).reshape(4, 4, 3)
print(arr, '\n')

value = arr[1, 0, -1]
print(value)
```

```python
arr = np.arange(1, 17).reshape(4, 4)
print(arr, '\n')

res = arr[1::2, 0::2]
print(res)
```

```python
arr = np.arange(1, 49).reshape(4, 4, 3)
print(arr, '\n')

sliced = arr[1, :2, :2]
print(sliced)
```

```python
arr = np.array([[23, 56, 78, 93], [71, 82, 13, 24]])
print(arr, '\n')

for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        if arr[i][j] % 2 != 0:
            arr[i][j] = -1

print(arr)
```

```python
arr = np.array([1, 0, 2, 0, 3, 0, 4])
print(arr)

res = np.nonzero(arr)
print(res[0])
```

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
print(arr1)
print(arr2)

add = arr1 + arr2
print("\nAddition:", add)
multiply = arr1 * arr2
print("Multiplication:", multiply)
```

```python
arr1 = np.array([15, 20, 25])
arr2 = np.array([10, 40, 37])
print(arr1)
print(arr2)

print(np.dot(arr1, arr2))
```
