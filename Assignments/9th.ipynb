```python
import pandas as pd

df = pd.DataFrame({
    "Date": ["2026-05-28", "2026-05-29", "2026-05-30"]
})

df["Date"] = pd.to_datetime(df["Date"])

df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Day_Name"] = df["Date"].dt.day_name()

print(df)
```
