```python
stu_name = input("Enter Name: ")
stu_class = input("Enter Class: ")
m1 = int(input("Marks in Physics (out of 100): "))
m2 = int(input("Marks in Chemistry (out of 100): "))
m3 = int(input("Marks in Maths (out of 100): "))
m4 = int(input("Marks in Hindi (out of 100): "))
m5 = int(input("Marks in English (out of 100): "))
total = m1 + m2 + m3 + m4 + m5
percentage = total / 5

print(f'''Name: {stu_name}
Class: {stu_class}
Total Marks: {total}
Percentage: {percentage:.2f}%''')
```

```python
str1 = str(input("Enter 1st string: "))
str2 = str(input("Enter 2nd string: "))
concat = str1 + " " + str2
print(f"Concatenated String: {concat}")
print(concat.lower())
print(concat.upper())
print(concat.title())
print(concat.swapcase())
print(concat.capitalize())
print(concat.casefold())
print(concat.center(50))
print(concat.count("l"))
print(concat.endswith("d"))
print(concat.find("o"))
print(concat.isalnum())
print(concat.isdigit())
print(concat.isnumeric())
print(concat.isspace())
print(concat.replace("o", "O"))
```
