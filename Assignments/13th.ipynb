```python
import numpy as np

arr = np.array([[6, -8, 73, -110], [np.nan, -8, 0, 94]])

arr = np.nan_to_num(arr, nan= 0)
print(arr, '\n')

arr = arr[::-1]
print(arr, '\n')

arr = arr[:, [3, 2, 1, 0]]
print(arr)
```

```python
import numpy as np

arr = np.arange(24).reshape(2, 3, 4)
print(arr.shape, '\n')
print(arr, '\n')

new_arr = np.moveaxis(arr, 0, 2)
print(new_arr.shape, '\n')
print(new_arr)
```

```python
import numpy as np

arr = np.array([[10, np.nan, 30], [20, 40, np.nan], [30, 50, 60]])

avg = np.nanmean(arr, axis=0)
arr = np.nan_to_num(arr, nan= avg)
print(arr)
```

```python
import numpy as np

arr = np.array([10, -5, 20, -8, 15])
arr[arr < 0] = 0
print(arr)
```

```python
import numpy as np
from statistics import mode

arr1 = np.array([[3, 4], [5, 6]])
arr2 = np.array([[1, 0], [7, 8]])

combined = np.concatenate((arr1, arr2))
print("Mean =", np.mean(combined))
print("Median =", np.median(combined))
print("Mode =", mode(combined.flatten()))
```

```python
import numpy as np

A = np.array([[1, -2, 3], [-1, 3, -1], [2, -5, 5]])
B = np.array([[9], [-6], [17]])

A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print("x =", X[0][0])
print("y =", X[1][0])
print("z =", X[2][0])
```

```python
import matplotlib.pyplot as plt

sub1 = ["Maths", "SPM", "DSA", "MEFA", "OS"]
sem1 = [75, 80, 78, 72, 85]
sub2 = ["Maths", "AI", "DBMS", "TOC", "CN"]
sem2 = [82, 85, 80, 79, 88]

plt.figure(figsize=(8, 5))
plt.plot(sub1, sem1, label="Semester 1")
plt.plot(sub2, sem2, label="Semester 2")

plt.title("Semester Result Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)
plt.show()
```
