```python
import csv

d = [
    ["Name", "Address", "Mobile", "Email"],
    ["Abc", "Jaipur", "939435", "abc@fe.fd"],
    ["Agsfr", "Kota", "23784", "abfec@fe.fd"],
    ["Flfw", "Kota", "8795", "vfeffc@fe.fd"],
    ["Qrdtf", "Alwar", "38762847", "ghsd@fe.fd"]
]

with open("addBook.csv", "w", newline="") as file:
    writer = csv.writer(file)
    
    for row in d:
        writer.writerow(row)
```

```python
import csv

d = []
name = input("Enter Name: ")
address = input("Enter Address: ")
mobile = input("Enter Mobile: ")
email = input("Enter Email: ")
d.append([name, address, mobile, email])

with open("addBook.csv", "a", newline="") as file:
    writer = csv.writer(file)
    
    for row in d:
        writer.writerow(row)
```

```python
import sqlite3
conn = sqlite3.connect("dbPractice.db")

sql = "create table one(one_id integer primary key autoincrement, name varchar(50), email varchar(50))"
conn.execute(sql)
sql = "create table two(two_id integer primary key autoincrement, item varchar(50), price integer)"
conn.execute(sql)

conn.close()
```

```python
import sqlite3
conn = sqlite3.connect("dbPractice.db")

sql = '''insert into one(name, email) values ("Alok", "alok@gmail.com"), ("Rahul", "rahul@gmail.com")'''
conn.execute(sql)
sql = '''insert into two(item, price) values ("Pen", 10), ("Book", 50)'''
conn.execute(sql)

conn.commit()
conn.close()
```

```python
import sqlite3
conn = sqlite3.connect("dbPractice.db")

sql = "select * from one"
res1 = conn.execute(sql)
sql = "select * from two order by item"
res2 = conn.execute(sql)

for x in res1:
    print(x)
for x in res2:
    print(x)
conn.close()
```
