```python
import numpy as np

arr = np.array([[6, -8, 73, -110], [np.nan, -8, 0, 94]])

arr = np.nan_to_num(arr, nan=0)
print(arr, '\n')

arr = arr.T
print(arr)
```

```python
import numpy as np

arr = np.arange(24).reshape(2, 3, 4)
print(arr.shape)
print(arr, '\n')

new_arr = np.moveaxis(arr, 0, 2)
print(new_arr.shape)
print(new_arr)
```

```python
import numpy as np

X = np.array([
    [1, 2, np.nan],
    [4, np.nan, 6],
    [7, 8, 9]
])
print(X, '\n')

col_avg = np.nanmean(X, axis=0)
inds = np.where(np.isnan(X))
X[inds] = np.take(col_avg, inds[1])
print(X)
```
