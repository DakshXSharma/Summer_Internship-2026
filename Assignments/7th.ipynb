```python
import pandas as pd

data1 = [1, 22, 333, 4444]
ps1 = pd.Series(data1)
print(ps1, type(ps1), '\n')

data2 = {'a': 1, 'b': 22, 'c': 333, 'd': 4444}
ps2 = pd.Series(data2)
print(ps2, type(ps2), '\n')

print(ps1[1])
print(ps2['c'])
```

```python
import pandas as pd

data1 = [[1, 2], [22, 33], [333, 444], [4444, 5555]]
df1 = pd.DataFrame(data1)
print(df1, type(df1), '\n')

data2 = {'a': [1], 'b': [22], 'c': [333], 'd': [4444]}
df2 = pd.DataFrame(data2)
print(df2, type(df2), '\n')

data3 = [[1, 2, 3], [22, 33, 44], [333, 444, 555], [4444, 5555, 6666]]
df3 = pd.DataFrame(data3)
print(df3, type(df3), '\n')

data4 = [(1, 2, 3), (22, 33, 44), (333, 444, 555), (4444, 5555, 6666)]
df4 = pd.DataFrame(data4)
print(df4, type(df4), '\n')

data5 = [{'a': 1, 'b': 22, 'c': 333, 'd': 4444}]
df5 = pd.DataFrame(data5)
print(df5, type(df5), '\n')
```

```python
import pandas as pd

data = {
    'Name': ['Aman', 'Riya', 'Rahul', 'Priya', 'Karan'],
    'Age': [20, 21, 19, 22, 20],
    'Marks': [85, 92, 78, 95, 88]
}
df = pd.DataFrame(data)
print(df, '\n')

print(df['Marks'], '\n')

print(df.loc[df['Name'] == 'Aman', 'Age'], '\n')

print(df[df['Marks'] > 85], '\n')

print(df.iloc[1], '\n')

print(df.loc[:2, 'Name'], '\n')

print(df.values.tolist())
```
