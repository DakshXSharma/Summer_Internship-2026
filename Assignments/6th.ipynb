```python
import requests

country = input("Enter country: ")
url = f"https://restcountries.com/v3.1/name/{country}"
response = requests.get(url)
data = response.json()
country_data = data[0]
print("Country:", country_data['name']['common'])
print("Capital:", country_data['capital'][0])
print("Population:", country_data['population'])
print("Region:", country_data['region'])
```

```python
import requests

pokemon = input("Enter Pokemon name: ")
url = f"https://pokeapi.co/api/v2/pokemon/{pokemon}"
response = requests.get(url)
data = response.json()
print("Name:", data['name'])
print("Height:", data['height'])
print("Weight:", data['weight'])
print("Base Experience:", data['base_experience'])
```
