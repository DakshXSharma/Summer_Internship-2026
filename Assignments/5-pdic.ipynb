```python
numbers = (12, 45, 7, 89, 23)

total = 0
highest = numbers[0]
lowest = numbers[0]

for i in numbers:
    total = total + i

    if i > highest:
        highest = i

    if i < lowest:
        lowest = i

print("Total:", total)
print("Highest:", highest)
print("Lowest:", lowest)
```

```python
n = (3, 14, 7, 22, 9, 41, 18, 5)

result = ()
for i in n:
    if i > 10:
        result = result + (i,)

print(result)
```
