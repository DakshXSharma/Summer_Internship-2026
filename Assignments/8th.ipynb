```python
import pandas as pd

df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Name": ["A1", "A2", "A3", "A4", "A5"],
    "Subject": ["sub1", "sub2", "sub4", "sub6", "sub8"]
})
df2 = pd.DataFrame({
    "ID": [1, 2, 3, 4, 7],
    "Name": ["B1", "B2", "B3", "B4", "B7"],
    "Subject": ["sub1", "sub2", "sub3", "sub4", "sub7"]
})

res1 = df1.merge(df2, on= 'ID', how= 'inner')
print(res1, '\n')

res2 = df1.merge(df2, on= 'ID', how= 'left')
print(res2, '\n')

res3 = df1.merge(df2, on= 'ID', how= 'right')
print(res3, '\n1')

res4 = df1.join(df2, on= 'ID', lsuffix= '_x', rsuffix= '_y')
print(res4)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Name": ["A1", "A2", "A3", "A4", "A5"],
    "Subject": ["sub1", "sub2", "sub4", "sub6", "sub8"]
})
df2 = pd.DataFrame({
    "ID": [1, 2, 3, 4, 7],
    "Name": ["B1", "B2", "B3", "B4", "B7"],
    "Subject": ["sub1", "sub2", "sub3", "sub4", "sub7"]
})
df3 = pd.DataFrame({
    "ID": [2, 3, 4, 6, 7],
    "Name": ["C2", "C3", "C4", "C6", "C7"],
    "Subject": ["sub1", "sub2", "sub3", "sub6", "sub7"]
})

res1 = pd.concat([df1, df2], ignore_index= True)
print(res1, '\n')

res2 = res1.merge(df3, on= 'ID')
print(res2)
```
