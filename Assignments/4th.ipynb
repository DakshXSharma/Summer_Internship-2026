```python
def maxOfThree(a, b, c):
    if(a >= b and a >= c):
        return a
    elif(b >= a and b >= c):
        return b
    else:
        return c

print(maxOfThree(55, 55, 4))
```

```python
def newList(l):
    distinct = []
    for i in l:
        if i not in distinct:
            distinct.append(i)
    return distinct

print(newList([2, 3, 2, 4, 3]))
```

```python
def mulList(*l):
    n = 1
    for x in l:
        n *= x
    return n

print(mulList(2, 72, 87, 5))
```

```python
def fact(n):
    if(n == 0):
        return 1
    else:
        return n * fact(n-1)

print(fact(5))
```

```python
def rev(st):
    return st[::-1]

print(rev("hi"))
```

```python
def func(s, e, num):
    if(num >= s and num <= e):
        return True
    else:
        return False

print(func(4, 10, 11))
```

```python
def evenNum(*l):
    for x in l:
        if(x % 2 == 0):
            print(x)

evenNum(1, 2, 3, 4, 5)
```

```python
def checkPrime(a):
    if(a <= 1):
        print("Not Prime!")
        return
    
    for i in range(2, a):
        if(a % i == 0):
            print("Not Prime!")
            return
    print("Prime!")

checkPrime(12)
```

```python
def strCase(a):
    upper = 0
    lower = 0
    for ch in a:
        if(ch.isupper()):
            upper += 1
        elif(ch.islower()):
            lower += 1
    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

strCase("HeLlo")
```
