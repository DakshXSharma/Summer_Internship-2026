```python
import pandas as pd

df = pd.DataFrame({
    'Col1': range(12),
    'Col2': ['A']*3 + ['B']*3 + ['C']*3 + ['D']*3,
    'Date': pd.to_datetime(['2026-05-27', '2026-05-28', '2026-05-29']*4)
})
print(df, '\n')

pivoted = df.pivot(index= 'Date', columns= 'Col2', values= 'Col1')
print(pivoted)
```

```python
import pandas as pd

data = {
    'Course': ['CS', 'Civil', 'CS', 'Civil', 'CS'],
    'Year': [1, 2, 1, 1, 2],
    'Student': [100, 150, 120, 90, 180]
}
df = pd.DataFrame(data)
print(df, '\n')

pivoted = df.pivot_table(index= 'Course', columns= 'Year', values= 'Student', aggfunc= 'sum')
print(pivoted)
```
