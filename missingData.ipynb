```python
import pandas as pd
import numpy as np

data = {
    'Col1': [3, np.nan, np.nan, 2],
    'Col2': [1, pd.NA, pd.NA, 4]
}
df = pd.DataFrame(data)
print(df, '\n')

f = df.fillna('-')
print(f, '\n')

res1 = df.ffill()
print(res1, '\n')

res2 = df.bfill()
print(res2)
```

```python
import pandas as pd
import numpy as np

data = [
        [1, 2, 3],
        [5, 6, 7],
        [34, 445, 98]
]
df = pd.DataFrame(data, index= ['a', 'c', 'd'], columns= ['one', 'two', 'three'])
print(df, '\n')

df = df.reindex(['a', 'b', 'c', 'd', 'e', 'f'])
print(df, '\n')

f = df.fillna("-")
print(f, '\n')

res = df.ffill(limit= 1)
print(res)
```

```python
import pandas as pd
import numpy as np

data = {'one': ['A', 'B', 'A', 'C', 'A', 'C', 'B'], 'two': [5, 6, 7, 78, 9, 54, 45]}
df = pd.DataFrame(data)
print(df, '\n')

gd = df.groupby('one')['two'].sum()
print(gd, '\n')
gd = df.groupby('one')['two'].mean()
print(gd, '\n')
gd = df.groupby('one')['two'].count()
print(gd, '\n')
gd = df.groupby('one')['two'].min()
print(gd, '\n')
gd = df.groupby('one')['two'].max()
print(gd)
```

```python
import pandas as pd
import numpy as np

data = {'one': ['A', 'B', 'A', 'C', 'A', 'C', 'B'], 'sc': ['X', 'Y', 'X', 'Z', 'Y', 'Z', 'Y'], 'two': [5, 6, 7, 78, 9, 54, 45]}
df = pd.DataFrame(data)
print(df, '\n')

res = df.groupby(['one', 'sc'])['two'].sum()
print(res, '\n')
```

```python
import pandas as pd

data = {'Name': ['ABC', 'XYZ', 'PQR'], 'Age': [28, 22, 34]}
df = pd.DataFrame(data)
print(df, '\n')

res = df.sort_values(by= 'Age', ascending= False)
print(res)
```

```python
import pandas as pd

data = {'Name': ['ABC', 'XYZ', 'PQR'], 'Age': [28, 22, 34]}
df = pd.DataFrame(data, [0, 2, 1])
print(df, '\n')

res = df.sort_index()
print(res)
```
