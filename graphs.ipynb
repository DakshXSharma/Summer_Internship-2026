```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.scatter(x, y)
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values, color= 'lightblue')
plt.title("Bar Graph")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = np.random.rand(1000)

plt.hist(data, bins= 10, color= 'lightblue', alpha= 0.8)
plt.title("Histo Graph")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```
